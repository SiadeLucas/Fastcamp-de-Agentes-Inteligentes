```python
# Definição da frase que queremos buscar
# junto com sua transformação em vetor
query_text = "vegetarian dishes"
query_vector = next(iter(model.embed(query_text)))


# O vetor da frase é comparado com os outros vetores guardados anteriormente
# para encontrar semelhança
results = client.query_points(
    collection_name="items",
    query=query_vector,
    with_payload=True,
    limit=5
)


# Os resultados são exibidos em ordem de relevância
# Quanto mais perto de 1.0, mais parecida é a frase
for result in results.points:
    print(f"Item: {result.payload.get('item_name', 'N/A')}")
    print(f"Score: {result.score}")
    print(f"Description: {result.payload['description'][:150]}...")
    print(f"Price: {result.payload.get('price', 'N/A')}")
    print("---")
```
